# Import required libraries

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pickle
```

# Image Preprocessing step by step

## Read Image Dataset
***
Digits hand written image dataset, created manually using **Wacom Tablet**

```python
image_path = os.getcwd() 
image_files = sorted([os.path.join(image_path, 'images', file) for file in os.listdir(image_path + "\\images") if file.endswith('.png')])
```

```python
image_files
```

```python
img = [cv2.imread(i, cv2.IMREAD_UNCHANGED) for i in image_files]
```

### Exploring image data

```python
for im in img:
    plt.imshow(im)
    plt.show()
```

## Resize image

```python
width = height = 80
dimensions = (width, height)
resized_images = list()

for i in range(len(img)):
    resized_image = cv2.resize(img[i], dimensions, interpolation=cv2.INTER_LINEAR)
    resized_images.append(resized_image)
```

```python
resized_images[0].shape
```

```python
for im in resized_images:
    plt.imshow(im)
    plt.show()
```

## Remove noise (noise reduction)
***
Remove noise using **Gaussian Blur** <br>
**GAUSSIAN BLUR (Gaussian Smoothing)** <br>
Melakukan *blurring* kepada gambar dengan menggunakan fungsi *Gaussian* dengan tujuan mengurangi noise dari image

```python
no_noise = []
for i in range(len(resized_images)):
    blur = cv2.GaussianBlur(resized_images[i], (5, 5), 0)
    no_noise.append(blur)
```

```python
for im in no_noise:
    plt.imshow(im)
    plt.show()
```

## Segmentation

```python
image = no_noise[2]
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```
